```python
import os

from tavily import TavilyClient

tavily_client = TavilyClient(api_key=os.getenv("TAVILY_API_KEY"))
response = tavily_client.search("오늘 한국 경제 관련 이슈 10가지?")
```

```python
import json

# Print a clear summary of the response
print("=== Tavily API Response Summary ===")
print(f"Query: {response.get('query')}")
print(f"Answer: {response.get('answer')}")
print(f"Follow-up questions: {response.get('follow_up_questions')}")
print(f"Images returned: {len(response.get('images', []))}")
print(f"Number of results: {len(response.get('results', []))}")
print(f"Response time: {response.get('response_time')}s")
print(f"Request ID: {response.get('request_id')}")
print("=" * 36 + "\n")

# Pretty-print the full response as JSON (readable, non-ASCII preserved)
print("Full response (JSON):")
print(json.dumps(response, indent=2, ensure_ascii=False))
print("\n" + "=" * 36 + "\n")

# Show a concise, readable list of results (if any)
results = response.get("results", [])

if results:
    print("Results (summarized):")
    for i, r in enumerate(results, start=1):
        title = r.get("title") or r.get("headline") or r.get("name") or r.get("source") or "<no title>"
        snippet = r.get("snippet") or r.get("summary") or r.get("excerpt") or ""
        url = r.get("url") or r.get("link") or "-"
        # Truncate snippet for readability
        snippet_short = (snippet[:300] + "...") if len(snippet) > 300 else snippet
        print(f"{i}. {title}")
        if snippet_short:
            print(f"   {snippet_short}")
        print(f"   URL: {url}\n")
else:
    print("No results returned.")
```
